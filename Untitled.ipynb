```python
from markov_crawler import MarkovCrawler
from graph_scorer import GraphScorer
import logging
import requests
from page_scorer import tokenize_alphanum
from bs4 import BeautifulSoup
import sys
```

```python
r = requests.get("http://search.ucas.com/subject/fulllist")
tokenized_sentences = []
soup = BeautifulSoup(r.text,"lxml")

for a in soup.find_all("a"):
    if "href" not in a.attrs:
        continue
    if not a.attrs["href"].startswith("/subject/"):
        continue
    tokenized_sentences.append(tokenize_alphanum(a.text))

g = GraphScorer(tokenized_sentences)
```

```python
# Set the logging level                                                                                                                                 
logging.getLogger("requests").setLevel(logging.WARNING)
logging.basicConfig(stream=sys.stderr,level=logging.DEBUG)


# Instantiate the crawler
raw_scores = []

urls = ["http://www.manchester.ac.uk",
        "https://www.masdar.ac.ae",
        "http://www.birmingham.ac.uk/index.aspx",
        "https://www.ajman.ac.ae/en",
        "https://www.sheffield.ac.uk",
        "https://www.adu.ac.ae",
        "https://aau.ac.ae",
        "https://www.aus.edu",
        "https://gmu.ac.ae",
        "http://www.buid.ac.ae","http://www.kustar.ac.ae",
        "http://nyuad.nyu.edu/en/","http://www.pi.ac.ae/en",
        "http://www.sharjah.ac.ae/en/Pages/default.aspx",
        "http://www.uowdubai.ac.ae/","http://www.zu.ac.ae/main/en/"]

for url in urls:
    with MarkovCrawler(max_depth=4,page_scorer=g.get_score,
                       threshold = 100,top_url=url) as crawler:
        n_tries = 0
        finished = False
        while not finished:
            crawler.prepare()
            finished = crawler.crawl()
            n_tries += 1
            if n_tries == 5:
                break
        raw_scores.append(crawler.raw_score)
```
